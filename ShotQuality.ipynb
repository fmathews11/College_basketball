```python
import requests,pandas as pd, numpy as np, json
```

```python
url = "https://shotquality.com/api/college_mens/2022/team-standings"
r = requests.get(url)
df = pd.json_normalize(r.json()['data'])
df.sort_values('NET_SQ',ascending = False,inplace = True)
df[df.Team_Name == "Purdue"].transpose()
```

```python
df.sort_values("NET_SQ",ascending = False)[['Team_Name','Offensive_Shot_Quality','Defensive_Shot_Quality','NET_SQ']].head(50)
```
